```python
class TareaGeneral:
    _tareas = []

    def __init__(self, nombre, descripcion, prioridad):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad

    def agregar_tarea(self):
        TareaGeneral._tareas.append(self)

    @staticmethod
    def listar_tareas():
        if TareaGeneral._tareas:
            print("Lista de tareas:")
            for i, tarea in enumerate(TareaGeneral._tareas, 1):
                print(f"{i}. {tarea.nombre} - {tarea.descripcion} ({tarea.prioridad})")
        else:
            print("No hay tareas en la lista.")

    @staticmethod
    def eliminar_tarea(nombre):
        for tarea in TareaGeneral._tareas:
            if tarea.nombre == nombre:
                TareaGeneral._tareas.remove(tarea)
                print(f"Tarea '{nombre}' eliminada correctamente.")
                return
        print(f"No se encontró ninguna tarea con el nombre '{nombre}'.")


class TareaFecha(TareaGeneral):
    def __init__(self, nombre, descripcion, prioridad, fecha):
        super().__init__(nombre, descripcion, prioridad)
        self.fecha = fecha

    @classmethod
    def listar_tareas(cls):
        # Llama al método de la clase base
        super().listar_tareas()
        for tarea in cls._tareas:
            if isinstance(tarea, TareaFecha):
                print(f"Fecha: {tarea.fecha}")


# Crear instancias de las tareas
tarea1 = TareaGeneral("Limpiar", "Hechar agua con lavandina el piso", "Baja")
tarea2 = TareaFecha(
    "Ordenar",
    "Ordenar toda ropa desordenada, cubiertos sueltos, etc",
    "Media",
    "2024-06-11",
)
tarea3 = TareaGeneral(
    "Saca la basura", "Antes que pase el camion de la basura hoy", "Alta"
)

# Agregamos las tareas
tarea1.agregar_tarea()
tarea2.agregar_tarea()
tarea3.agregar_tarea()

# Listar todas las tareas y mostrar
TareaGeneral.listar_tareas()
```

```python
class TareaGeneral:
    def __init__(self, nombre, descripcion, prioridad):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad

    def __str__(self):
        return f"Nombre: {self.nombre}, Descripción: {self.descripcion}, Prioridad: {self.prioridad}"


class TareaConFecha:
    def __init__(self, nombre, descripcion, prioridad, fecha):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad
        self.fecha = fecha

    def __str__(self):
        return f"Nombre: {self.nombre}, Descripción: {self.descripcion}, Prioridad: {self.prioridad}, Fecha: {self.fecha}"


class ListaTareas:
    def __init__(self):
        self.tareas = []

    def agregar_tarea(self, tarea):
        self.tareas.append(tarea)

    def eliminar_tarea(self, nombre):
        for tarea in self.tareas:
            if tarea.nombre == nombre:
                self.tareas.remove(tarea)
                print(f"Tarea '{nombre}' eliminada correctamente.")
                return
        print(f"No se encontró ninguna tarea con el nombre '{nombre}'.")

    def listar_tareas(self):
        if self.tareas:
            print("Lista de tareas:")
            for i, tarea in enumerate(self.tareas, 1):
                print(f"{i}. {tarea}")
        else:
            print("No hay tareas en la lista.")


# Ejemplo de uso
tareas = ListaTareas()

tarea1 = TareaGeneral("Limpiar", "Limpiar la casa", "Alta")
tarea2 = TareaConFecha("Comprar", "Comprar víveres", "Media", "2024-06-10")
tarea3 = TareaGeneral("Ordenar", "Ordenar la pieza", "Alta")

tareas.agregar_tarea(tarea1)
tareas.agregar_tarea(tarea2)
tareas.agregar_tarea(tarea3)

tareas.listar_tareas()

tareas.eliminar_tarea("Comprar")
```
