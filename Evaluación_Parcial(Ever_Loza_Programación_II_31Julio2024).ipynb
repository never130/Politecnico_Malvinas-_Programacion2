<u>Ejercicio N° 1: </u>  
Imagina que estás desarrollando un programa para una máquina expendedora de bebidas. Implementar un programa que solicite al usuario seleccionar una opción y muestre el precio correspondiente de la bebida seleccionada.

Implementa una solución que cumpla con los siguientes requisitos:

Muestra al usuario un menú con las siguientes opciones de bebidas:

+ Opción 1: Coca-Cola - Precio: $2.50
+ Opción 2: Sprite - Precio: $2.00
+ Opción 3: Fanta - Precio: $1.75
Solicita al usuario que seleccione una opción ingresando el número correspondiente.

Verificar la opción seleccionada por el usuario y mostrar por pantalla el precio correspondiente.

Si el usuario selecciona una opción inválida, mostrar un mensaje de error.




```python
# Ejercicio 1
def mostrar_menu():
    """Muestra el menú de bebidas y sus precios."""
    print("\nBienvenido a la máquina de bebidas!!!")
    print("Seleccione una opción:")
    print("1: Coca-Cola - Precio: $2.50")
    print("2: Sprite - Precio: $2.00")
    print("3: Fanta - Precio: $1.75")
    print("4: Salir")  # Opción para salir del programa


def obtener_bebida_y_precio(numero):
    """
    Retorna el nombre y el precio de la bebida según la opción seleccionada.

    Parámetros:
    numero (int): El número de opción seleccionada.

    Retorna:
    tupla: Una tupla con el nombre de la bebida y el precio, o (None, None) si la opción no es válida.
    """
    bebidas = {1: ("Coca-Cola", 2.50), 2: ("Sprite", 2.00), 3: ("Fanta", 1.75)}

    if numero in bebidas:
        return bebidas[
            numero
        ]  # Si la opción es válida, retorna la tupla con nombre y precio
    else:
        return None, None  # Si la opción no es válida, retorna una tupla (None, None)


# Ejecucion del programa, con un bucle while hasta que el usuario desee salir del bucle de menu
while True:
    mostrar_menu()
    try:
        # Solicita al usuario que ingrese una opción
        numero = int(input("Ingrese el número de la opción que desea: "))
        if numero == 4:
            print("Gracias por usar la máquina de bebidas. Nos vemos!!")
            break  # Salir del bucle y finalizar el programa

        # Llama a la función para obtener el nombre y el precio de la bebida
        nombre_bebida, precio = obtener_bebida_y_precio(numero)
        if nombre_bebida:
            print(f"Bebida seleccionada: {nombre_bebida}")
            print(f"El precio de la bebida seleccionada es: ${precio}")
        else:
            # Muestra un mensaje de error si la opción es inválida
            print("Opción inválida. Por favor, seleccione una opción válida.")

    except ValueError:
        # Maneja la excepción si el usuario ingresa algo que no es un número
        print("Error: Por favor, ingrese un número válido.")
    except KeyboardInterrupt:
        # Maneja la excepción si el usuario interrumpe el programa (Ctrl+C)
        print("\nInterrupción del usuario. El programa ha sido terminado.")
        break
    except Exception as e:
        # Maneja cualquier otra excepción que no haya sido considerada
        print(f"Ha ocurrido un error inesperado: {e}")
```

<u>Ejercicio N° 2 : </u>  
Imagina que estás creando un programa para una biblioteca que gestiona la información de los libros que tiene disponibles. Tu tarea es implementar un sistema de gestión de libros que permita a los usuarios realizar consultas y obtener información sobre los libros.   
Utilizando los conceptos de Programación Orientada a Objetos, implementa una solución que cumpla con los siguientes requisitos:
1.	Crea una clase llamada Libro que tenga los siguientes atributos:
    + titulo: una cadena que representa el título del libro.
	+ autor: una cadena que indica el autor del libro.
	+ precio: un número decimal que indica el precio del libro.
	+ disponible: un valor booleano que indica si el libro está disponible para préstamo.
2.	La clase Libro debe tener los siguientes métodos:
	+ informacion(): muestra por pantalla la información del libro, incluyendo el título, autor y precio.
	+ prestar(): cambia el estado del libro a no disponible.
	+ devolver(): cambia el estado del libro a disponible.
3.	Crea una clase derivada llamada LibroFisico que herede de la clase Libro. La clase LibroFisico tiene un atributo adicional:
	+ ubicacion: una cadena que indica la ubicación física del libro en la biblioteca.
4.	Sobrescribe el método informacion() en la clase LibroFisico para incluir la ubicación física del libro al mostrar la información.



```python
# Ejercicio 2
class Libro:
    def __init__(self, titulo, autor, precio):
        """Iinicializa un nuevo objeto de tipo libro
        parametros,
        titulo (str), el titulo del libro
        autor (str), el autor del libro
        precio (float) el precio del libro
        disponible (boolean): indica si el libro esta disponible
        """
        self.titulo = titulo
        self.autor = autor
        self.precio = precio
        self.disponible = True

    # imprime la informacion del libro
    def informacion(self):
        print(f"Titulo: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Precio: {self.precio}")

    # Este metodo cambia mediante una condicion if else, si esta prestado
    def prestar(self):
        if self.disponible:
            self.disponible = False
            print("El libro ha sido prestado")
        else:
            print("El libro no esta prestado")

    # Este metodo cambia mediante una condicion if else, si esta devuelto
    def devolver(self):
        if not self.disponible:
            self.disponible = True
            print("El libro ha sido devuelto")
        else:
            print("El libro no esta devuelto")

# Esta clase hereda atributos de la clase padre(Libro) y se le agrega un atributo adicional
class Libro_Fisico(Libro):
    def __init__(self, titulo, autor, precio, ubicacion):
        super().__init__(titulo, autor, precio)
        self.ubicacion = ubicacion

    def informacion(self):
        super().informacion()
        print(f"Ubicacion: {self.ubicacion}")


# Ejemplo de uso
# El bloque try intenta ejecutar el codigo dentro de el si ocurre un error, lanza una excepcion y se imprime el error.
try:
    print("Libro: ")
    # Se crean las instancias de libro y libro_fisico y se llaman a sus metodos para mostrar informacion.
    libro_1 = Libro("Harry Potter", "JK Rowling", 30.30)
    libro_1.informacion() # Se imprime la informacion del libro
    libro_1.prestar() # Se imprime el metodo de prestar libro
    libro_1.informacion() # Se vuelve a mostrar el libro con su estado actual.
    print("\nLibro Fisico: ")
    # Se crea una instancia de la clase Libro_fisico
    libro_2 = Libro_Fisico(
        "Estructura de datos", "John Connor", 50.99, "Area Programacion"
    )
    libro_2.informacion() # Se imprime la informacion obtenida del libro
    libro_2.prestar()
    libro_2.devolver()
    libro_2.prestar()
    libro_2.informacion() # Se vuelve a imprimir el estado actual del libro


except Exception as e:
    print("Error al ejecutar el codigo de las clases", e)
```
